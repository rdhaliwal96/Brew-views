```python
# Import dependencies
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
from decimal import Decimal
```

```python
# Connection
engine = create_engine("mysql://root:password@localhost/brewery_db")
conn = engine.connect()

# Query db
city_data = pd.read_sql("SELECT * FROM us_city_data", conn)

# Manipulate coordinates col
latitude = [coord.split(',')[0] for coord in city_data['Coordinates']]
longitude = [coord.split(',')[1] for coord in city_data['Coordinates']]
city_data['latitude'] = latitude
city_data['longitude'] = longitude

city_data.head()
```

```python
# Send updates to db
for index, row in city_data.iterrows():
    print(f"Updating {row.City} longitude and latitude")    
    engine.execute(f'UPDATE us_city_data SET latitude = {row.latitude},  longitude = {row.longitude} WHERE city = "{row.City}"')
```

```python
# Query top 11 west coast cities
top_cities = pd.read_sql("SELECT * FROM brewery_db.us_city_data WHERE longitude < -117.1611 ORDER BY population DESC", conn)
cities_list = list(top_cities['City'][0:10])
cities_list
```
